```python
# you need to import a library that can handle random data through random functions, make random choices, generate random numbers, etc.
import random

# num_trials is the argument, you can plug in different numbers to simulate games and see whether you have a better probability of win a car by switching or staying
def monty_hall_simulation(num_trials):
    
    # gives the score, whether or not you win a car, when you switch or stay
    switch_wins = 0
    stay_wins = 0

    for _ in range(num_trials):
        # Step 1: create a list for the doors
        doors = ['goat', 'goat', 'car']
        
        # Step 2: Shuffle the doors so that the goats and car are place randomly behind doors
        random.shuffle(doors)

        # Step 3: Contestant makes an initial choice out of the three doors
        contestant_choice = random.choice([0, 1, 2])

        # Step 4: Host opens one of the other doors (from the two doors that the contestant didn't pick)
        available_doors = [i for i in range(3) if i != contestant_choice and doors[i] == 'goat']
        host_opened = random.choice(available_doors)

        # Step 5: Determine the remaining door the contestant can switch to if he decides to switch
        remaining_door = [i for i in range(3) if i != contestant_choice and i != host_opened][0]

        # Step 6: Assign a way to determine if switching or staying gives the contestant a better probabilit of winning a car
        if doors[remaining_door] == 'car':
            switch_wins += 1
        if doors[contestant_choice] == 'car':
            stay_wins += 1
  
    # returns the score
    return switch_wins, stay_wins

# Run the simulation for a large number of trials
num_trials = 10000
switch_wins, stay_wins = monty_hall_simulation(num_trials)

# Calculate probabilities
print(f"Switching wins: {switch_wins} ({(switch_wins / num_trials) * 100:.2f}%)")
print(f"Staying wins: {stay_wins} ({(stay_wins / num_trials) * 100:.2f}%)")
```
